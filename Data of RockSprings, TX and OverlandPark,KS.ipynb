```python
url = "http://www.century21.com/real-estate/rocksprings-tx/LCTXROCKSPRINGS"
r = requests.get(url)
con = r.content
soup = BeautifulSoup(con,"html.parser")
#print(soup.prettify())
all_data = soup.find_all("div",{"class":"property-card"}) #Extracting "div" tags
l = []
for item in all_data:
    d = {}
    #print(item)
    d["Price"]= item.find("a",{"class":"listing-price"}).text.replace("\n","").replace(" ","")
    d["Address"] = item.find("div",{"class","property-address"}).text.replace("\n","").replace("  ","")
    d["City"] = item.find("div",{"class","property-city"}).text.replace("\n","").replace("  ","")
    try:
        d["Beds"] = item.find("div",{"class","property-beds"}).text.replace("\n","")
    except:
        d["Beds"] = None
        
    try:
        d["Bath"] = item.find("div",{"class","property-baths"}).text.replace("\n","")
    except:
        d["Bath"] = None
        
    try:
        d["SQFT"] = item.find("div",{"class","property-sqft"}).text.replace("\n","")
    except:
        d["SQFT"] = None 
    
    l.append(d)    
```

```python
df = pd.DataFrame(l)
df
```
